```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Data collection and prepossing
sonar_data=pd.read_csv("/content/Copy of sonar data.csv",header=None)
sonar_data.head()
```

```python
#shape i.e., no. of rows and columns
sonar_data.shape
```

```python
#statistical definitions for the sonar_data
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#seperate the data and label
x=sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
print(x)
print(y)
```

```python
#splitting the data into thetraing part and testing part
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
print(x_train)
print(y_train)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
acc_predict=accuracy_score(pred,y_test)
```

```python
print("accuracy on test data:", acc_predict)
```
